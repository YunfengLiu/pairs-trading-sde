```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import real as rl
from simulation import portfolio_process
```

## MSFT vs PFE

```python
# Five years of close data from MSFT and PFE
msft_close_data = yf.download("MSFT", start="2017-03-01", end="2022-02-28")["Close"].to_numpy()
pfe_close_data = yf.download("PFE", start="2017-03-01", end="2022-02-28")["Close"].to_numpy()
```

```python
T = 5
# use 251 as number of trading days
dt = 1 / 251
B = msft_close_data
S = np.log(B)
A = pfe_close_data
X = np.log(A) - np.log(B)
```

```python
np.corrcoef(A, B)
```

```python
# plot stock pairs
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(B, label="MSFT")
ax.plot(A, label="PFE")
ax.legend()
ax.set_title("MSFT, PFE stock prices")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(X, label="Spread X")
ax.legend()
ax.set_title("MSFT, PFE Spread X")
plt.show()
```

```python
sigma_h = rl.sigma_hat(S, dt)
mu_h = rl.mu_hat(S, dt)
k_h = rl.k_hat(X, dt)
theta_h = rl.theta_hat(X)
eta_h = rl.eta_hat(X, dt)
rho_h = rl.rho_hat(X, S, dt)
```

```python
V0 = 1000
gamma = -100
r = 0.02
pv, h_weights = portfolio_process(A, B, X, V0, r, T, k_h, gamma, eta_h, sigma_h, rho_h, theta_h)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(h_weights, label="Portfolio Weights")
ax.legend()
ax.set_title("MSFT and PFE h* over time")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(pv, label="Portfolio value")
ax.legend()
ax.set_title("MSFT and PFE Wealth Dynamic")
plt.show()
```
